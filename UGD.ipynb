## Tipe A:

1. Confidence score di kiri atas bounding box.
2. Confidence score dan bounding box menjadi biru tua.
3. Warning text merah

### Kunci Tipe A

```python
import cv2
import numpy as np

# Memuat model deteksi wajah pre-trained
model_file = "deploy.prototxt"
weights_file = "res10_300x300_ssd_iter_140000.caffemodel"
face_net = cv2.dnn.readNetFromCaffe(model_file, weights_file)

# Mengambil input video dari kamera
video_capture = cv2.VideoCapture(0)

def detect(frame):
    (h, w) = frame.shape[:2]

    # Mengubah gambar menjadi blob
    blob = cv2.dnn.blobFromImage(frame, 1.0, (300, 300), (104.0, 177.0, 123.0))
    face_net.setInput(blob)
    
    # Melakukan deteksi wajah
    detections = face_net.forward()

    faces_detected = False  # Flag untuk memeriksa apakah wajah terdeteksi

    # Proses deteksi wajah untuk setiap hasil
    for i in range(0, detections.shape[2]):
        confidence = detections[0, 0, i, 2]

        # Hanya deteksi wajah dengan confidence lebih dari 0.5
        if confidence > 0.5:
            faces_detected = True  # Wajah terdeteksi

            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x_start, y_start, x_end, y_end) = box.astype("int")

            # Menambahkan kotak biru di sekitar wajah yang terdeteksi
            cv2.rectangle(frame, (x_start, y_start), (x_end, y_end), (255, 0, 0), 2)
            
            # Menambahkan teks untuk confidence score dengan warna biru
            label = f"{confidence * 100:.2f}%"
            cv2.putText(frame, label, (x_start, y_start - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
    
    # Jika tidak ada wajah yang terdeteksi, tampilkan warning text berwarna merah di tengah jendela
    if not faces_detected:
        text = "Wajah tidak terdeteksi"
        (text_width, text_height), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 1.0, 2)
        
        # Menghitung posisi teks agar berada di tengah
        x_text = (w - text_width) // 2
        y_text = (h + text_height) // 2
        
        # Menampilkan teks merah di tengah jendela
        cv2.putText(frame, text, (x_text, y_text), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2)

    return frame

while True:
    result, video_frame = video_capture.read()

    if not result:
        break 

    video_frame = detect(video_frame)

    # Menampilkan frame dengan deteksi wajah dan confidence score
    cv2.imshow("Real-Time Face Detection with Confidence Score", video_frame)

    # Keluar dari loop jika tombol 'q' ditekan
    if cv2.waitKey(1) & 0xFF == ord("q") or cv2.getWindowProperty("Real-Time Face Detection with Confidence Score", cv2.WND_PROP_VISIBLE) < 1:
        break

# Menutup video capture dan windows setelah selesai
video_capture.release()
cv2.destroyAllWindows()
```

### Kunci Tipe D
